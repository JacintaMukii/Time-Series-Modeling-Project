### 4.6 Creating the ROI and CV columns 

In order to develop a more comprehensive approach to address the objectives outlined for this project, we will incorporate two additional columns into our dataset: ROI(Return on Investment) and CV (Coefficient of Variation), respectively. These columns will provide valuable insights and enhance our analytical capabilities in pursuit of our project goals.

 - #### Return on Investment  (ROI):
   ROI serves as an indicator of the anticipated returns from investments.It is a crucial metric for real estate investors as it provides a quantitative measure of the profitability and performance of a real estate investment project. It helps investors make informed decisions, assess risk, and compare different investment opportunities in the dynamic and potentially lucrative real estate market.

 - #### Coefficient of Variation (CV):
   CV acts as a gauge of data point dispersion relative to the mean. It quantifies the ratio of the standard deviation to the mean, assisting investors in evaluating the level of volatility or risk in relation to the expected return on investments. 

```python
# calculating and creating a new column -ROI

df['ROI'] = (df['2018-04']/ df['1996-04'])-1



#calculating std to be used to find CV
df["std"] = df.loc[:, "1996-04":"2018-04"].std(skipna=True, axis=1)

#calculating mean to be used to find CV
df["mean"] = df.loc[:, "1996-04":"2018-04"].mean(skipna=True, axis=1)

# calculating and creating a new column - CV

df["CV"] = df['std']/df["mean"]

# dropping std and mean as they are not necessary for analysis

df.drop(["std", "mean"], inplace=True, axis=1)
```

```python
#Removing Outliers using standard deviation
def determine_outlier_thresholds_std(Numeric, col_name):
    upper_boundary = Numeric[col_name].mean() + 3 * Numeric[col_name].std()
    lower_boundary = Numeric[col_name].mean() - 3 * Numeric[col_name].std()
    return lower_boundary, upper_boundary
```

```python
def check_outliers_std(Numeric, col_name):
    lower_boundary, upper_boundary = determine_outlier_thresholds_std(Numeric, col_name)
    if Numeric[(Numeric[col_name] > upper_boundary) | (Numeric[col_name] < lower_boundary)].any(axis=None):
        return True
    else: 
        return False

def replace_with_thresholds_std(Numeric, cols, replace=False):
    from tabulate import tabulate
    data = []
    for col_name in cols:
        if col_name != 'Outcome':
            outliers_ = check_outliers_std(Numeric, col_name)
            count = None
            lower_limit, upper_limit = determine_outlier_thresholds_std(Numeric, col_name)
            if outliers_:
                count = Numeric[(Numeric[col_name] > upper_limit) | (Numeric[col_name] < lower_limit)][col_name].count()
                if replace:
                    if lower_limit < 0:
                        # We don't want to replace with negative values, right!
                        Numeric.loc[(Numeric[col_name] > upper_limit), col_name] = upper_limit
                    else:
                        Numeric.loc[(Numeric[col_name] < lower_limit), col_name] = lower_limit
                        Numeric.loc[(Numeric[col_name] > upper_limit), col_name] = upper_limit
            outliers_status = check_outliers_std(Numeric, col_name)
            data.append([outliers_, outliers_status,count, col_name, lower_limit, upper_limit])
    table = tabulate(data, headers=['Outlier (Previously)','Outliers','Count', 'Column','Lower Limit', 'Upper Limit'], tablefmt='rst', numalign='right')
    print("Removing Outliers using 3 Standard Deviation")
    print(table)

replace_with_thresholds_std(eda_df, Numeric,replace=True)    
```

```python
# Plot histograms of the cleaned dataset after removing the outliers
# Set the font size for the titles
title_fontsize = 16
tick_label_fontsize = 8 

# Loop through numeric columns and create histograms
for column in Numeric:
    plt.figure(figsize=(12, 6))  # Set the figure size
    plt.hist(eda_df[column], bins=70)
    
    # Add title and set font size
    plt.title(column, fontsize=title_fontsize)
    
    # Customize axis labels if needed
    plt.xlabel("X-axis Label", fontsize=10)
    plt.ylabel("Frequency", fontsize=12)
    # Rotate x-axis labels for better readability (optional)
    plt.tight_layout()
    
    plt.grid(False)  # Add grid lines if desired
    plt.show()
```

### 7.1 Components of Time Series

In time series analysis, a time series data set is typically decomposed into four main components. These components help in understanding and modeling the underlying patterns and variations in the data. The four components are:

   - ##### Trend (T):
           The trend component represents the long-term movement or behavior of the time series 
           data. It captures the overall direction in which the data is moving over time. Trends 
           can  be ascending, descending, or relatively stable.

   - ##### Seasonality (S): 
           Seasonality refers to repeating and predictable patterns or fluctuations in the data 
           that occur at regular intervals, often associated with calendar seasons, months, weeks,
           or other periodic occurrences. Seasonal patterns are usually driven by external factors 
           and are not part of the underlying trend.

   - ##### Cyclic Patterns (C): 
           The cyclic component represents non-seasonal fluctuations that occur over a more 
           extended period than seasonal patterns. Unlike seasonality, cyclic patterns do not have 
           fixed or predictable durations, and they are often influenced by economic, 
           environmental, or business cycles.

   - ##### Irregular or Residual (I): 
           The irregular component, also known as the residual component or noise, includes all 
           the random and unpredictable fluctuations in the data that cannot be attributed 
           to the trend, seasonality, or cyclic patterns. It represents the variability and noise 
           that is inherent in most real-world data.

These four components are often combined in an additive or multiplicative manner, depending on the nature of the data and the specific modeling approach. Decomposing a time series into these components allows analysts and data scientists to better understand the underlying structure of the data, identify patterns, and make more accurate forecasts or predictions.

```python
# checking for trend, seasonality and residuals in data
def seasonal_decomposition(df):
    decomposition = seasonal_decompose(df)
    
    # Gather the trend, seasonality, and residuals 
    trend = decomposition.trend
    seasonal = decomposition.seasonal
    residual = decomposition.resid

# Plot gathered statistics
    plt.figure(figsize=(10,8))
    plt.subplot(411)
    plt.plot(df, label='Original', color='red')
    #plt.plot(ts, label='Original', color='blue')
    plt.legend(loc='best')
    plt.subplot(412)
    plt.plot(trend, label='Trend', color='red')
    plt.legend(loc='best')
    plt.subplot(413)
    plt.plot(seasonal,label='Seasonality', color='red')
    plt.legend(loc='best')
    plt.subplot(414)
    plt.plot(residual, label='Residuals', color='red')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.xticks(fontsize =12)
    
```

### 7.2  Stationarity

We will utilize the Dickey-Fuller test to ascertain the non-stationary nature of the data.

```python
def check_stationarity(df):
    result = adfuller(df)
  
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
```

```python
check_stationarity(series_2)
```

The time series is confirmed to be non-stationary because both the p-value and the test statistic exceed their respective critical values, with the p-value being greater than 0.05 and the test statistic being greater than the critical values.

### 7.3 Differencing

Differencing is a  technique in time series analysis used to make non-stationary data suitable for modeling. It simplifies the data by removing trends, stabilizing variance, and facilitating the identification of underlying patterns and relationships in the time series.

```python
def detrend(df):
    '''function returns a stationary series'''
    
    diff_series = df.diff(1).diff(12).dropna()
    return diff_series
```

```python
# removing trend from the series  using the function above 
detrended_series = detrend(series_2)
```

```python
#check the stationarity of the detrended column
check_stationarity(detrended_series)
```

With a p-value lower than 0.05(alpha) and a test statistic below all critical values, we can confidently conclude that the series has been transformed into a stationary state.

### 7.5 Time Series Modelling

Upon recognizing the presence of seasonality within the data, it becomes apparent that an appropriate modeling approach should account for these recurring patterns. SARIMAX, a versatile modeling technique, emerges as an ideal choice. SARIMAX not only captures the underlying time trends but also incorporates the seasonality component, allowing for a more comprehensive representation of the data's temporal dynamics. This makes it a robust option for modeling the observed patterns effectively.

```python
#using auto_arima- it does a random search for the best pdq,PDQS  
sarima_model = pm.auto_arima(series_2, 
                             m=12,
                             seasonal=True,
                             start_p=0,
                             start_q=0,
                             start_P=0,
                             start_Q=0,
                             max_order=6, 
                             test='adf',
                             error_action='warn',  
                             suppress_warnings=True,
                              stepwise=True,
                              trace=False)
```

* When evaluating and comparing statistical models fitted with different parameters, each can be ranked against one another based on how well it fits the data or its ability to accurately predict future data points
 
* The Akaike Information Criterion (AIC) tests the goodness of fit.It rewards models that achieve a high goodness-of-fit with little complexity
* A model that fits the data very well while using lots of features will be assigned a larger AIC score than a model that uses fewer features to achieve the same goodness-of-fit. 
* Auto_arima does a random search and comes up with the best parameters(from the given ones) that reduce the AIC.

```python
#Displaying the summary of the model
sarima_model.summary()
```

This model has a high AIC. Will look for other values of pdq and s that might lower it

```python
# Define the p, d and q parameters to take any value between 0 and 2

p = d = q = range(0,2)
```

```python
# Generate all different combinations of p, d , q and s
pdq = list(itertools.product(p, d, q))
pdqs = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
ans = []
for comb in pdq:    
    for combs in pdqs:
        try:
            model = sm.tsa.statespace.SARIMAX(series_2,
                                            order=comb,
                                            seasonal_order=combs,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            output = model.fit()
            ans.append([comb, combs, output.aic])
           
        except:
            continue
```

```python
ans_df = pd.DataFrame(ans, columns=['pdq','pdqs', 'aic'])
ans_df
```

In this method, we perform in-sample prediction. The model makes one-step-ahead forecasts by utilizing the actual value from the preceding time step, rather than relying on a predicted value.
